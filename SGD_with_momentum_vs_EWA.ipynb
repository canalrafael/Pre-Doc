```python
import numpy as np
import matplotlib.pyplot as plt

# Initial configuration
num_iterations = 100  # Number of iterations
np.random.seed(3)    # Seed for reproducibility

# Parameters for SGD with Momentum
mu = 0.01               # Momentum coefficient
epsilon = 0.1          # Learning rate

# Parameters for EWA Filter
q = 1 - mu             # Correspondence with mu
alpha = epsilon        # Correspondence with epsilon

# Initializing variables
theta_sgd = 0.0        # Initialization of theta for SGD with momentum
v_sgd = 0.0            # Initialization of momentum v

theta_ewa = 0.0        # Initialization of theta for EWA
v_ewa = 0.0            # Initialization of momentum v

# Store the trajectory of theta for comparison
trajectory_sgd = []
trajectory_ewa = []

# Perform updates for num_iterations iterations
for t in range(num_iterations):
    # Simulated stochastic gradient
    grad_t = np.random.randn()  # Noisy gradient

    # Update for SGD with Momentum
    v_sgd = mu * v_sgd + grad_t
    theta_sgd -= epsilon * v_sgd
    trajectory_sgd.append(theta_sgd)

    # Update for EWA Filter
    v_ewa = (1 - q) * v_ewa + q * grad_t
    theta_ewa -= alpha * v_ewa
    trajectory_ewa.append(theta_ewa)

# Plot the trajectories
plt.figure(figsize=(10, 6))
plt.plot(trajectory_sgd, label='SGD with Momentum', linestyle='--', color='blue')
plt.plot(trajectory_ewa, label='EWA Filter', linestyle='-', color='red')
plt.title('Comparison between SGD with Momentum and EWA Filter')
plt.xlabel('Iterations')
plt.ylabel('Value of $\\theta_t$')
plt.legend()
plt.grid(True)
plt.show()
```
